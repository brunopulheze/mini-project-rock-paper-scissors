```python
import random

print("""
Winning rules of the game Rock Paper and Scissors are:
Rock vs Paper -> Paper wins
Rock vs Scissors -> Rock wins
Paper vs Scissors -> Scissors wins
"""
)

# Variable to control if loop is active as True
is_game_running = True

# Score list
score_list = {"user" : 0, "computer": 0}

while is_game_running:
    # Takes user input as strings 'Rock', 'Paper', 'Scissors' or 'Quit' to end the game
    choice = input("Enter your choice: Rock, Paper, Scissors or Quit to leave the game")
    print(f"User entered {choice}")

    # If user's input is 'Quit' it changes loop control to False
    if choice.lower() == 'quit':
        if score_list["user"] > score_list["computer"]:
            print(f"The winner is user!")
        elif score_list["user"] < score_list["computer"]:
            print(f"The winner is computer!")
        else:
            print(f"The game is draw!")
        is_game_running = False

    # Translate user input from string into integer
    def translate_user_input(s):
        user_input = 0
        if s.lower() == 'rock':
            user_input = 1
        elif s.lower() == 'paper':
            user_input = 2
        elif s.lower() == 'scissors':
            user_input = 3
        else: 
            print("Input does not exist")
        
        print(f"User input {s} was translated as: {user_input}")
        return user_input
            
    # Save the result of the function as a variable to be used in the next function to compare user's and computer's choices
    users_choice = translate_user_input(choice)

    # Calculate the score function
    def calculate_score(string):
            for key, value in score_list.items():
                print(f"Score Dictionary initial state is Key: {key}, Value: {value}")
                if string == 'computer':
                    score_list['computer'] += 1
                elif string == 'user':
                    score_list['user'] +=1
            
    # Function to compare the user's and computer's choices
    def compare_choices():
        print(f"User's choice {users_choice}")
        # Create a random number for the computer's choice
        computers_choice = random.randint(1, 3)
        print(f"Computer's choice {computers_choice}")
        
        if users_choice == computers_choice:
            print(f"Draw - Both user and computer had {users_choice}")
            
        elif users_choice == 1 and computers_choice == 2:
            print(f"Computer wins - Paper wins")
            calculate_score('computer')
        elif users_choice == 1 and computers_choice == 3:
            print(f"User wins - Rock wins")
            calculate_score('user')
        elif users_choice == 2 and computers_choice == 3:
            print(f"User wins - Scissor wins")
            calculate_score('user')
        elif computers_choice == 1 and users_choice == 2:
            print(f"Computer wins - Paper wins")
            calculate_score('computer')
        elif computers_choice == 1 and users_choice == 3:
            print(f"Computer wins - Rock wins")
            calculate_score('computer')
        elif computers_choice == 2 and users_choice == 3:
            print(f"User wins - Scissor wins")
            calculate_score('user')
        
    compare_choices()
    
    # Print score list
    print(f"""Score: {score_list}
            ____________""")
```
